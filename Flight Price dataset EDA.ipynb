### Feature Engineering

Handling "Date_of_Journey" column

```python
# Splitting Date Month and Year seperately

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# Converting Date Month and year to numeric datatype

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop(columns='Date_of_Journey', inplace=True) # Dropping the 'Date_of_Journey' column
```

```python
df.head()
```

Handling "Arrival_Time" column

```python
# Keeping only time and removing date 

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# Splitting hour and min seperately

df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
# Converting Date Month and year to numeric datatype

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.head()
```

```python
df.drop(columns = 'Arrival_Time', inplace=True) # Dropping the 'Arrival_Time' column
```

```python
df.head()
```

Handling "Dep_Time" column

```python
# Splitting hour and min seperately

df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
# Converting Date Month and year to numeric datatype

df['Dep_min'] = df['Dep_min'].astype(int)
df['Dep_hour'] = df['Dep_hour'].astype(int)
```

```python
df.drop(columns = 'Dep_Time', inplace = True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.rename(columns = {'Dep_hour' : 'Departure_hour', 'Dep_min' : 'Departure_min'}, inplace=True) # Renaming the columns 'Dep_hour' and 'Dep_min'
```

```python
df.head(2)
```

Handling "Route" column

```python
df.drop(columns = 'Route', inplace=True) # Dropping the 'Route' column
```

Handling Duration column

```python
# Extract the hour part from the 'Duration' column
# Example: '2h 50m' → ['2h', '50m'] → '2h' → '2'

df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
# Extract the minute part from the 'Duration' column
# Example: '2h 50m' → ['2h', '50m'] → '50m' → '50'

df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head(5)
```

```python
# Replace missing values (NaN) in the 'Duration_min' column with 0
# This is useful for rows like '3h' where there is no minute part, which would result in NaN

df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
# Convert the 'Duration_hour' column to string type
# This is useful before using string methods like .str.replace()

df['Duration_hour'] = df['Duration_hour'].astype(str)

# Convert the 'Duration_min' column to string type
# Ensures consistent data type before further string operations

df['Duration_min'] = df['Duration_min'].astype(str)
```

```python
# Clean the 'Duration_hour' column:
# - Remove 'h' if it's still present
# - Remove 'm' in case minute values were mistakenly placed in hour column
# - Replace any empty strings ('') with '0' to avoid errors during conversion

df['Duration_hour'] = df['Duration_hour'].str.replace('h', '').str.replace('m', '').replace('', '0')

# Clean the 'Duration_min' column:
# - Remove 'h' in case hour values ended up here
# - Remove 'm' to keep only the numeric part
# - Replace empty strings with '0' for safety

df['Duration_min'] = df['Duration_min'].str.replace('h', '').str.replace('m', '').replace('', '0')
```

```python
# Fill any missing values (NaN) in the 'Duration_hour' column with '0'
# This ensures there are no empty/missing hour values before converting to integers

df['Duration_hour'] = df['Duration_hour'].fillna('0')

# Fill any missing values (NaN) in the 'Duration_min' column with '0'
# Useful for cases where minutes were missing (like durations written as '3h' only)

df['Duration_min'] = df['Duration_min'].fillna('0')
```

```python
# Convert the cleaned 'Duration_hour' column from string to integer
# This makes it suitable for mathematical operations like addition or modeling

df['Duration_hour'] = df['Duration_hour'].astype(int)

# Convert the cleaned 'Duration_min' column from string to integer
# Ensures you can calculate total duration or use it as a numerical feature

df['Duration_min'] = df['Duration_min'].astype(int)
```

```python
df.drop(columns = 'Duration', inplace = True)
```
